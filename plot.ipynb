```python
def pwm_para_v(pwm):
    return (pwm/255)*5

def carregar_dados(arquivo):
    tempo = []
    valor = []

    with open(arquivo, 'r') as f:
        for linha in f:
            t, v = linha.strip().split(',')
            tempo.append(int(t))
            valor.append(float(v))
        return tempo, valor
```

```python
import matplotlib.pyplot as plt
  
tempo_pwm, pwm = carregar_dados('pwm_linear.txt')
tempo_rpm, rpm = carregar_dados('rpm_linear.txt')

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(tempo_pwm, pwm, 'r-')
plt.title('Entrada PWM (Linear)')
plt.xlabel('Amostras')
plt.ylabel('PWM')

plt.subplot(2, 1, 2)
plt.plot(tempo_rpm, rpm, 'b-')
plt.title('Sada RPM (Linear)')
plt.xlabel('Amostras')
plt.ylabel('RPM')

plt.tight_layout()
plt.show()
```
